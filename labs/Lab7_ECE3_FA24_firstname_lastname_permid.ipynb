Q1. Compute and Plot the FFT of a Sine Wave  
Write a Python program to:  
1. Generate a sine wave with a frequency of 100 Hz, sampled at 1000 points over a duration of 1 second.  
2. Use `scipy.fft` to compute the Fast Fourier Transform (FFT) of the signal.  
3. Plot both the original sine wave and its FFT magnitude spectrum.  

Sample Output:
- The first plot should show the sine wave in the time domain.  
- The second plot should show peaks in the frequency domain corresponding to the sine wave's frequency.  
image.png



```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt

points = 1000

space = np.linspace(0, 1, points)
sin = np.sin(2 * np.pi * 100 * space)

yf = sci.fft.fft(sin)
xf = sci.fft.fftfreq(points, 1/points)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

ax1.set_title("Time-Domain Signal")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Amplitude")
ax1.plot(space, sin, "blue")

ax2.set_title("Frequency-Domain Signal (FFT Magnitude Spectrum)")
ax2.set_xlabel("Frequency (Hz)")
ax2.set_ylabel("Magnitude")
ax2.set_xlim(0, 500)
ax2.plot(xf, yf)

fig.tight_layout()
```

Q2. Solving a System of Linear Equations

Write a Python program that solves the following system of linear equations:

$2x + 3y = 8$ \\
$x - y = 1$

#### Requirements:

1. Use **NumPy** to represent the system as matrix equations in the form $AX = B$, where:
   - $A$ is the coefficient matrix.
   - $X$ is the vector of variables $[x, y]$.
   - $B$ is the vector of constants $[8, 1]$.

2. Solve for $X$ using `np.linalg.solve()` and print the values of $ x $ and $ y $.

#### Expected Output:

The program should print the values of $x$ and $y$ that satisfy the system of equations.


```python
import numpy as np

A = [[2, 3], [1, -1]]
b = [8, 1]

b = np.linalg.solve(A, b)
print(f"X = {b[0]}\nY = {b[1]}")
```
